```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
constituents = pd.read_csv('sp500_constituents.csv', sep = ",")
```

```python
#Pandas a mal lu le fichier. Ainsi, par exemple, le symbole "BRKb" est devenu "BRK.B". Nous allons donc modifier cela
Symbols = constituents['Symbol'].values
for i in range(len(Symbols)):
    Symbols[i] = Symbols[i].replace('.','')
```

```python
Symbols_df = pd.DataFrame(Symbols)
```

```python
constituents['Symbol'] = Symbols_df
```

```python
#Liste des liens. Nous ignorons dans un premier temps les différences ".N" ou ".OQ" à la fin
link = 'https://www.reuters.com/finance/stocks/financial-highlights/'
Links = []
for i in range(len(Symbols)):
    Links.append(link + Symbols[i] + '.N')
```

```python
def extract(name,link):

    req = requests.get(link)
    soup = BeautifulSoup(req.text, "lxml")
    
    ta= soup.find_all('table')
    # Future liste des en-tetes
    L=[]
    # Future liste des valeurs
    R=[]

    #Autres tables
    for p in range(2,9) :
        t=ta[p]
        table_rows = t.find_all('tr') 
        for tr in table_rows :
            th=tr.find_all('th')
            td=tr.find_all('td')
            if len(th) !=0 :
                rowH=[i.get_text(strip=True) for i in th]
                
            if len(td) !=0 :
                row =[i.get_text(strip=True) for i in td]
                row[0] = row[0].replace('\xa0',' ')
                row=row[0:2]
                for j in range(1,len(row)):
                    row[j] = row[j].replace(",","")
                    try:
                        row[j] = float(row[j])
                    except:
                        if row[j] == '--':
                            row[j] = None
                
                if len(td)!=1 :
                    l=[]
                    for i in range(1,2) :
                        l=l+[row[0] + ' (' +rowH[i]+')']
                    res = row[1:]
        
                    L=L+l
                    R=R+res
                    
    #Ajout de la derniere table                
 
    #Ajout de la premiere table  

    #Ajout de la colonne avec les noms
    L= ["Nom de l'entreprise"] +L  
    R=[name]+R
    E= L,R 
    return E
```

```python
#Renvoie les valeurs et les en-tetes pour chaque entreprise
#Dense et dur a lire mais précis
V=[]
L,N= Links, Symbols
n=len(L)
    
for i in range(n):
    print("Ecriture de ",N[i],'|| progression : ',i,'/',n)
        #Toutes les cotes des entreprise ne sont en XXX.N mais parfois en XXX.OQ : si on ne récupere rien avec le premier lien, on teste avec le second
    try :
        EL,ER=extract(N[i],L[i])
        V=V+ [EL,ER]
        print('done : Entreprise de catégorie 1')
        print(len(EL), len(ER))
    except :
        L[i]=(L[i]).replace('.N','.OQ')
        EL,ER=extract(N[i],L[i])
        V=V +[EL,ER]
        print('done : Entreprise de catégorie 2')
        print(len(EL), len(ER))
```

```python
m = len(V)//2
features = []
data = []
for i in range(0,m,2):
    head = V[i]
    for j in range(len(head)):
        if head[j] not in features:
            features.append(head[j])
```

```python
f = len(features)
for i in range(m):
    head = V[2*i]
    data.append([None]*f)
    for feature in features:
        if feature in head:
            indiceh = head.index(feature) #indice dans head
            indicef = features.index(feature) #indice dans features
            data[i][indicef] = V[2*i+1][indiceh]
```
